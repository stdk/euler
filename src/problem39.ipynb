```python
p,x,y = symbols('p,x,y')
f = (p - x - y)**2
g = f.expand() - x*x - y*y
g.subs(x,(-root(2,2)+2)*p/2).subs(y,(-root(2,2)+2)*p/2).simplify()
```

```python
plot_implicit(Or(Eq(x-60*(2+root(2,2))),Eq(y-60*(2+root(2,2))),Eq(x-y,0),Eq(g.subs(p,120),0)),(x,-50,250),(y,-50,250))
```

```python
h = (2*p*x-p*p)/(2*x-2*p)
```

```python
solutions = {}
for kp in range(1,1001):
    fy = h.subs(p,kp)
    for kx in range(1,int(kp*(2-root(2,2))/2)+1):
        ky = fy.subs(x,kx)
        if type(ky) == Integer:
            if solutions.get(kp) is None:
                solutions[kp] = 0
            solutions[kp] += 1
            #print kx,ky,root(kx*kx+ky*ky,2)
print max(solutions.iterkeys(), key=(lambda key: solutions[key]))
```
